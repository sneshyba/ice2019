```python
# Generic imports
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
%matplotlib notebook
```

```python
pwd = os.getcwd()
startpos = pwd.find('crystals/')
crystal = pwd[startpos+9:]; print(crystal)
```

```python
case41 = np.loadtxt('case2.0 (calibration)/Segments2roughnessZ2.txt')
Z2_case41 = case41[:,0]
logrhoZ2_case41 = case41[:,1]

case49 = np.loadtxt('case2.1/Segments2roughnessZ2.txt')
Z2_case49 = case49[:,0]
logrhoZ2_case49 = case49[:,1]

case411 = np.loadtxt('case2.2/Segments2roughnessZ2.txt')
Z2_case411 = case411[:,0]
logrhoZ2_case411 = case411[:,1]
```

```python
plt.figure()
markersize = 7

plt.plot(Z2_case41,logrhoZ2_case41,'og',markersize=markersize)
thisrange = [i for i in range(0,8)]
p1 = np.polyfit(Z2_case41[thisrange],logrhoZ2_case41[thisrange],1)
best_sigma1 = 1/np.sqrt(-p1[0])
print(p1,'predicted sigma = ', best_sigma1)
label_sigma1 = 'case 2.0, $\sigma$='+str(np.round(best_sigma1*100)/100) 
plt.plot(Z2_case41,np.polyval(p1,Z2_case41),'g', label=label_sigma1)

plt.plot(Z2_case49,logrhoZ2_case49,'ob',markersize=markersize)
thisrange = [i for i in range(0,8)]
p1 = np.polyfit(Z2_case49[thisrange],logrhoZ2_case49[thisrange],1)
best_sigma1 = 1/np.sqrt(-p1[0])
print(p1,'predicted sigma = ', best_sigma1)
label_sigma1 = 'case 2.1, $\sigma$='+str(np.round(best_sigma1*100)/100) 
plt.plot(Z2_case49,np.polyval(p1,Z2_case49),'b', label=label_sigma1)

plt.plot(Z2_case411,logrhoZ2_case411,'ok',markersize=markersize)
thisrange = [i for i in range(0,13)]
p1 = np.polyfit(Z2_case411[thisrange],logrhoZ2_case411[thisrange],1)
best_sigma1 = 1/np.sqrt(-p1[0])
print(p1,'predicted sigma = ', best_sigma1)
label_sigma1 = 'case 2.2, $\sigma$='+str(np.round(best_sigma1*100)/100) 
plt.plot(Z2_case411,np.polyval(p1,Z2_case411),'k', label=label_sigma1)

plt.grid(True)
plt.title(crystal)
plt.xlabel('$Z^2$')
plt.ylabel(r'$ln(\rho)$')
plt.legend()
plt.xlim([.01,.1])
plt.ylim([-2.7,-.6])
```
